```python
import pandas as pd
import numpy as np
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
airports_ex = pd.read_csv('airports-extended.csv')
```

```python
airports_ex.head()
```

```python
airports_ex.columns = ['Airport ID', 'Name', 'City', 'Country', 'IATA', 'ICAO', 'Latitude', 'Longitude', 'Altitude', 'Timezone', 'DST', 'Tz database time zone','Type', 'Source']
```

```python
airports_ex.head()
```

```python
airports_ex['Type'].value_counts()
```

```python
airports =  airports_ex.drop(airports_ex.loc[airports_ex['Type'] != 'airport' ].index)
```

```python
airports['Type'].value_counts()
```

```python
airports.info()
```

```python
airports['Country'].value_counts()
```

```python
plot_df = (
    airports['Country'].value_counts()
        .head(10)
        .rename_axis('Country')
        .reset_index(name='no_of_airports')
)
plot_df
```

```python
x_labels = plot_df['Country']
fig = plot_df.plot(kind='bar')

fig.set_xticklabels(x_labels)
fig.bar_label(fig.containers[0], label_type='edge')
```

```python
plot_df1 = (
    airports['Country'].value_counts()
        .tail(27)
        .rename_axis('Country')
        .reset_index(name='no_of_airports')
)
plot_df1
```

```python
plt.figure(figsize = (15,5))
plt.bar(plot_df1['Country'], plot_df1['no_of_airports'])
```
